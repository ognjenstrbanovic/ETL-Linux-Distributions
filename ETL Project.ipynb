```python
import pandas as pd
import json
import pymongo
```

```python
df = pd.read_csv("Distrowatch_popularity.csv")
df.head()
```

```python
new_df = df.drop(columns = ["Unnamed: 0"])
```

```python
final_df = new_df.rename(columns={"name": "Name", "12_months": "12_Months", "6_months": "6_Months", "3_months": "3_Months", "1_month": "1_Month"})
```

```python
json_loads = json.loads(final_df.to_json(orient='records'))
```

```python
json_loads
```

```python
connection = 'mongodb://localhost:27017'
client = pymongo.MongoClient(connection)
db = client.etlProject
```

```python
collection = db.linux
```

```python
collection.remove()
collection.insert(json_loads)
print(collection.count())
```

```python
returned_data = db.linux.find().sort('_id', -1)
for i in range(10):
    print(returned_data[i])
```
